```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import time

# 1. 크롤러(Driver) 관리 클래스
class WebDriverManager:
    def __init__(self):
        options = Options()
        options.add_argument('--headless=new')
        options.add_argument('--no-sandbox')
        options.add_argument('--disable-dev-shm-usage')
        self.driver = webdriver.Chrome(options=options)

    def quit(self):
        self.driver.quit()
```

```python


class LG_PDF_URL:
    def __init__(self, driver, url):
        self.driver = driver
        self.url = url
        self.wait = WebDriverWait(self.driver, 20)

    def open(self):
        self.driver.get(self.url)
        time.sleep(1)

    def click_tab(self):
        try:
            # 'data-link-name'이 'support'인 a 태그 직접 찾기
            pdf_url_a = self.driver.find_element(By.CSS_SELECTOR, "div.contents-menu-list a[data-link-name='support']")
            self.driver.execute_script("arguments[0].scrollIntoView(true);", pdf_url_a)
            self.driver.execute_script("arguments[0].click();", pdf_url_a)
            time.sleep(1)
            return True
        except Exception as e:
            print(f"    ❌ 리뷰 탭 클릭 실패: {e}")
            return False

    def get_pdf_url(self):
        try:
            self.open()  # 상세페이지 열기
            # 제품 사용설명서가 들어간 a 태그 모두 찾기
            pdf_links = self.driver.find_elements(By.XPATH, "//a[contains(@href, 'https://gscs-b2c.lge.com/downloadFile?fileId=')]")
            for link in pdf_links:
                text = link.text.strip()
                href = link.get_attribute('href')
                if href.startswith('https://gscs-b2c.lge.com/downloadFile?fileId='):
                    return href
            print("    ❌ PDF 링크를 찾을 수 없음")
            return None
        except Exception as e:
            print(f"    ❌ PDF 링크 수집 실패: {e}")
            return None


```

```python
# 결과를 여기에 저장

lg_pdf_url = []
ll=list(df['product_url'].unique())
print(ll)
print(len(ll))
```

```python
# WebDriver 설정
manager = WebDriverManager()
driver = manager.driver
```

```python

for url in ll:
    try:
        print(f"🔍 크롤링 중: {url}")
        page = LG_PDF_URL(driver, url)
        page.open()
        url_srcs = page.get_pdf_url()
        lg_pdf_url.append(url_srcs)
        print(f"   👉 수집 PDF파일링크: {(url_srcs)}")
    except Exception as e:
        print(f"   ❌ 에러 발생: {e}")
        lg_pdf_url.append([])
print(lg_pdf_url)

```
